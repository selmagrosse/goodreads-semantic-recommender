```python
import pandas as pd

df = pd.read_csv("../data/goodreads_library_cleaned_descriptions.csv")
```

```python
df["Description"].isna().sum()
```

# Data cleaning

```python
# Enforce string type and fill NaNs with empty strings
df['Description'] = df['Description'].astype(str).fillna("")
```

```python
# Strip leading/trailing whitespace
df['Description'] = df['Description'].str.strip()
```

```python
df[["Title", "Author", "Description"]].head(20)
```

```python
# Lowercase all text
df['Description'] = df['Description'].str.lower()
```

```python
# Replace repeated whitespace with single space
df['Description'] = df['Description'].str.replace(r'\s+', ' ', regex=True)
```

```python
df[["Title", "Author", "Description"]].head(20)
```

```python
# Remove URLs
df['Description'] = df['Description'].str.replace(r'http\S+|www\.\S+', '', regex=True)
```

```python
print(df['Description'].head(20))
```

```python
# Replace repeated punctuation with single punctuation, excluding dots
df['Description'] = df['Description'].str.replace(r'([!?,;:-])\1+', r'\1', regex=True)
```

```python
# Normalize multiple spaced dots to standard ellipsis
df['Description'] = df['Description'].str.replace(r'(\.\s){2,}\.', '...', regex=True)
```

```python
 # Remove non-printable / OCR junk
df['Description'] = df['Description'].str.replace(r'[\x00-\x1f\x7f-\x9f]', '', regex=True)
```

```python
# Strip again after replacements to clean up any leftover spaces
df['Description'] = df['Description'].str.strip()
```

```python
print(df['Description'].head(20))
```

```python
df.to_csv("../data/goodreads_library_descriptions_cleaned.csv", index=False)
```
