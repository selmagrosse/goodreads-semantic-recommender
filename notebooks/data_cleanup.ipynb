```python
# Keep Title, Author, category, rating, and average rating columns
goodreads_df = goodreads_df[['Title', 'Author', 'ISBN', 'ISBN13', 'Number of Pages', 'Year Published', 'Original Publication Year', 'Bookshelves', 'Exclusive Shelf', 'My Rating', 'Average Rating']]
```

```python
goodreads_df.head()
```

```python
goodreads_df[goodreads_df['Number of Pages'].isna()]
```

```python
# Add has_page_count column
goodreads_df['Number of Pages'] = pd.to_numeric(goodreads_df['Number of Pages'], errors='coerce')   
goodreads_df['has_page_count'] = goodreads_df['Number of Pages'].notna().astype(int)
```

```python
goodreads_df[goodreads_df['Number of Pages'].isna()]
```

```python
goodreads_df[goodreads_df['Year Published'] != goodreads_df['Original Publication Year']]
```

```python
goodreads_df = goodreads_df.drop(columns=['Year Published'])
```

```python
goodreads_df.columns
```

```python
goodreads_df[goodreads_df['Original Publication Year'].isna()]
```

```python
goodreads_df['Original Publication Year'].isna().sum()
```

```python
goodreads_df[goodreads_df['Bookshelves'].isna()]
```

```python
# Fill NaNs in 'Bookshelves' with 'read' if 'Exclusive Shelf' is 'read'
goodreads_df.loc[(goodreads_df['Bookshelves'].isna()) & (goodreads_df['Exclusive Shelf'] == 'read'), 'Bookshelves'] = 'read'
```

```python
goodreads_df['Bookshelves'].isna().sum()
```

```python
goodreads_df.isna().sum()
```

```python
goodreads_df.head()
```

```python
# Check for duplicate Title and Author combinations
dups = goodreads_df[goodreads_df.duplicated(subset=["Title", "Author"], keep=False)]
dups.sort_values(["Title", "Author"])
```

```python
# Remove duplicates keeping the most complete entry
goodreads_df = (goodreads_df.sort_values(by=["Bookshelves", "My Rating"],
                                         ascending=[True, True]
                                         ).drop_duplicates(subset=["Title", "Author"], keep="first")
)
```

```python
goodreads_df[goodreads_df.duplicated(subset=['Title', 'Author'])]
```

```python
# Clean ISBN and ISBN13 columns
goodreads_df["ISBN"] = goodreads_df["ISBN"].astype(str).str.strip().str.replace('=', '', regex=False).str.replace('"', '', regex=False)
goodreads_df["ISBN13"] = goodreads_df["ISBN13"].astype(str).str.strip().str.replace('=', '', regex=False).str.replace('"', '', regex=False)
```
